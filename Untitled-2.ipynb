```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 한글 폰트 설정 (환경에 따라 'Malgun Gothic' 또는 'AppleGothic' 사용)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

# 1. 12월 전체 데이터 합산 (AS-IS)
# 데이터 프레임 생성 (컬럼: 상황, 인덱스: 감정, 값: 1~4/5주차 합산)
# (이미지_0~3의 데이터를 수동으로 합산하여 반영하는 것이 가장 정확합니다.)
# 여기서는 시뮬레이션을 위해 이미지_4의 4~5주차 데이터를 기반으로 전체적인 경향성을 반영한 가상의 전체 데이터를 사용합니다.
# (실제 데이터로 교체 가능)

situations = ['대출 탐색', '대출 신청', '심사 결과', '앱 UX', '고객 지원', '혜택 이벤트', '사후 관리']
sentiments = ['만족 기대', '감사/기쁨', '불안', '불만', '분노', '아이러니']

# 가상의 12월 전체 데이터 (AS-IS)
# (이미지_4의 데이터를 기반으로 전체적인 경향성을 반영하여 값을 증폭)
entire_dec_asis = np.array([
    [80, 0, 5, 0, 0, 0],  # 대출 탐색 (만족 중심)
    [30, 0, 10, 5, 0, 0], # 대출 신청 (불안/불만 일부)
    [5,  0, 10, 15, 15, 5],# 심사 결과 (부정 감정 다수)
    [10, 0, 5,  40, 20, 5],# 앱 UX (불만/분노 집중)
    [15, 0, 5,  10, 5,  0], # 고객 지원
    [20, 0, 0,  0,  0,  0], # 혜택 이벤트
    [40, 5, 30, 10, 5,  5]  # 사후 관리 (불안/불만 다수)
]).T

df_asis = pd.DataFrame(entire_dec_asis, index=sentiments, columns=situations)

# 2. TO-BE 시나리오 적용 (12월 전체 개선)
# - 개선 우선순위 전략 반영 (앱UX/심사결과/사후관리 부정 감정 70~80% 감소)
# - 감소된 건수의 80%를 '만족 기대'로 전이
df_tobe = df_asis.copy()

# 개선 함수 (부정 감정을 만족으로 전이)
def apply_improvement(df, situation, sentiment_from, sentiment_to='만족 기대', reduction_rate=0.8, transition_rate=0.8):
    count_from = df.loc[sentiment_from, situation]
    count_reduced = int(count_from * reduction_rate)
    count_transitioned = int(count_reduced * transition_rate)
    
    df.loc[sentiment_from, situation] -= count_reduced
    df.loc[sentiment_to, situation] += count_transitioned

# 시나리오 적용
apply_improvement(df_tobe, '앱 UX', '불만', reduction_rate=0.8, transition_rate=0.8) # 불만 80% 감소, 만족으로 80% 전이
apply_improvement(df_tobe, '앱 UX', '분노', reduction_rate=0.9, transition_rate=0.8) # 분노 90% 감소 (더 강력하게 제어), 만족으로 80% 전이
apply_improvement(df_tobe, '심사 결과', '분노', reduction_rate=0.8, transition_rate=0.8)
apply_improvement(df_tobe, '심사 결과', '불만', reduction_rate=0.7, transition_rate=0.8)
apply_improvement(df_tobe, '사후 관리', '불안', reduction_rate=0.8, transition_rate=0.8)
apply_improvement(df_tobe, '대출 신청', '불안', reduction_rate=0.6, transition_rate=0.8)

# 3. 시각화: AS-IS vs TO-BE 비교 히트맵 (색상 농도 방식)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(22, 10))

# AS-IS: 기존처럼 진한 파란색 계열
sns.heatmap(df_asis, annot=True, fmt='d', cmap='YlGnBu', ax=ax1, cbar_kws={'label': 'VOC 건수'})
ax1.set_title('AS-IS: 개선 전 12월 전체 VOC 분포', fontsize=18)
ax1.set_xlabel('상황 카테고리', fontsize=14)
ax1.set_ylabel('감성 카테고리', fontsize=14)
ax1.tick_params(axis='x', rotation=45)

# TO-BE: 개선된 곳은 색상이 옅어지거나 새로운 연두색으로 표현
# - 기존 색상 척도를 그대로 사용하되, 건수가 줄어들면 색상이 옅어지는 효과를 활용
# - 긍정 감정으로 이동한 건수들은 새로운 연두색 척도를 사용하여 시각적으로 분리
sns.heatmap(df_tobe, annot=True, fmt='d', cmap='YlGnBu', ax=ax2, cbar_kws={'label': 'VOC 건수'})
ax2.set_title('TO-BE: 프로세스 개선 후 시뮬레이션 (12월 전체)', fontsize=18, color='green')
ax2.set_xlabel('상황 카테고리', fontsize=14)
ax2.set_ylabel('감성 카테고리', fontsize=14)
ax2.tick_params(axis='x', rotation=45)

# (참고: TO-BE 그래프에서 긍정 감정으로 전이된 건수들을 더 명확히 보여주려면, 
#  긍정/부정을 분리하여 다른 색상 척도를 적용하는 것이 좋습니다.)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 한글 폰트 설정
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

situations = ['대출 탐색 비교', '대출 신청', '심사 결과', '앱 UX', '고객 지원', '혜택 이벤트', '사후 관리']
sentiments = ['만족 기대', '감사/기쁨', '불안', '불만', '분노']

# 1. 2월 1~4주차 데이터 합산 (AS-IS)
# 이미지에서 분석된 2월 전체 누적 수치
feb_asis_data = np.array([
    [91, 11, 29, 31, 0],   # 대출 탐색 비교
    [46, 8, 74, 60, 10],   # 대출 신청
    [15, 2, 60, 80, 22],   # 심사 결과
    [8, 0, 23, 108, 9],    # 앱 UX
    [22, 27, 34, 43, 9],   # 고객 지원
    [25, 3, 0, 2, 0],      # 혜택 이벤트
    [9, 5, 2, 7, 0]        # 사후 관리
]).T

df_feb_asis = pd.DataFrame(feb_asis_data, index=sentiments, columns=situations)

# 2. TO-BE 시나리오 적용 (80% 개선 및 만족 전이)
df_feb_tobe = df_feb_asis.copy()

def apply_improvement(df, situation, sentiment_from_list):
    for sent_from in sentiment_from_list:
        count = df.loc[sent_from, situation]
        reduced = int(count * 0.8)
        df.loc[sent_from, situation] -= reduced
        df.loc['만족 기대', situation] += reduced

# 2월 타겟: 대출 신청, 심사 결과, 앱 UX의 모든 부정 감정 개선
targets = ['대출 신청', '심사 결과', '앱 UX']
for sit in targets:
    apply_improvement(df_feb_tobe, sit, ['불안', '불만', '분노'])

# 3. 시각화 (Grayscale 적용)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(22, 10))

# 데이터 최대값에 맞춰 명암 대비 조절
v_max = df_feb_asis.values.max()

# AS-IS 히트맵 (Greys 색상 적용)
sns.heatmap(df_feb_asis, annot=True, fmt='d', cmap='Greys', ax=ax1, vmax=v_max, cbar_kws={'label': 'VOC 건수'})
ax1.set_title('AS-IS: 개선 전 2월 전체 VOC 분포', fontsize=18, fontweight='bold', pad=20)
ax1.set_xlabel('상황 카테고리', fontsize=12)
ax1.set_ylabel('감성 카테고리', fontsize=12)

# TO-BE 히트맵 (Greys 색상 적용)
sns.heatmap(df_feb_tobe, annot=True, fmt='d', cmap='Greys', ax=ax2, vmax=v_max, cbar_kws={'label': 'VOC 건수'})
ax2.set_title('TO-BE: 프로세스 개선 후 시뮬레이션 (2월 전체)', fontsize=18, fontweight='bold', pad=20, color='#333333')
ax2.set_xlabel('상황 카테고리', fontsize=12)
ax2.set_ylabel('감성 카테고리', fontsize=12)

plt.tight_layout()
plt.show()

# 4. 점수 계산 출력
def calc_score(df):
    s_w = {'만족 기대': 1, '감사/기쁨': 1, '불안': 4, '불만': 3, '분노': 5}
    i_w = {'대출 탐색 비교': 2, '대출 신청': 2, '심사 결과': 4, '앱 UX': 5, '고객 지원': 3, '혜택 이벤트': 2, '사후 관리': 3}
    total = 0
    for sit in df.columns:
        for sent in df.index:
            total += df.loc[sent, sit] * s_w[sent] * i_w[sit]
    return total

score_asis = calc_score(df_feb_asis)
score_tobe = calc_score(df_feb_tobe)

print(f"2월 전체 AS-IS 점수: {score_asis}")
print(f"2월 전체 TO-BE 점수: {score_tobe}")
print(f"개선 효율: {(1 - score_tobe/score_asis)*100:.1f}% 리스크 감소")
```
