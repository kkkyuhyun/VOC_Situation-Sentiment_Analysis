```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
import platform

# ── 한글 폰트 설정 ──────────────────────────────────────────────
def set_korean_font():
    system = platform.system()
    if system == "Darwin":  # macOS
        plt.rcParams['font.family'] = 'AppleGothic'
    elif system == "Windows":
        plt.rcParams['font.family'] = 'Malgun Gothic'
    else:  # Linux (Colab 등)
        # 나눔폰트 설치 필요: apt-get install fonts-nanum
        plt.rcParams['font.family'] = 'NanumGothic'
    plt.rcParams['axes.unicode_minus'] = False

set_korean_font()

# ── 데이터 ──────────────────────────────────────────────────────
감성 = ["만족", "기대", "감사·기쁨", "불안", "불만", "분노", "아이러니"]
상황 = ["대출탐색·비교", "대출신청", "심사·결과", "앱UX", "고객지원", "혜택·이벤트", "사후관리"]

# 행=감성, 열=상황 (전체 데이터 기준)
data = np.array([
    [10,  3, 28,  1,  0,  0,  4],  # 만족
    [ 0,  0,  2,  0,  1,  0,  1],  # 기대
    [ 2,  0, 28,  0,  0,  0,  6],  # 감사·기쁨
    [ 0,  0,  1,  0,  0,  0,  0],  # 불안
    [10,  4,  5,  5,  2,  4,  1],  # 불만
    [ 2,  2,  4,  0,  1,  1,  0],  # 분노
    [ 0,  0,  1,  0,  0,  0,  0],  # 아이러니
])

# ── 히트맵 그리기 ─────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(12, 6))

# 색상맵: 긍정(초록)·부정(빨강) 계열 분리를 위해 YlOrRd 사용
cmap = plt.cm.YlOrRd
im = ax.imshow(data, cmap=cmap, aspect='auto')

# 축 레이블
ax.set_xticks(range(len(상황)))
ax.set_yticks(range(len(감성)))
ax.set_xticklabels(상황, fontsize=11, rotation=20, ha='right')
ax.set_yticklabels(감성, fontsize=11)

# 감성 레이블 색 구분 (긍정/부정)
긍정_감성 = {"만족", "기대", "감사·기쁨"}
for label in ax.get_yticklabels():
    label.set_color("#1A6B3C" if label.get_text() in 긍정_감성 else "#C0392B")

# 셀 안에 숫자 표시
max_val = data.max()
for i in range(len(감성)):
    for j in range(len(상황)):
        val = data[i, j]
        text_color = "white" if val > max_val * 0.6 else "black"
        text = str(val) if val > 0 else "-"
        ax.text(j, i, text, ha='center', va='center',
                fontsize=12, fontweight='bold', color=text_color)

# 컬러바
cbar = fig.colorbar(im, ax=ax, shrink=0.8, pad=0.02)
cbar.set_label("VOC 건수", fontsize=10)

# 제목 및 레이아웃
ax.set_title("VOC 감성 × 상황 교차 히트맵 (전체 데이터)",
             fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel("상황 카테고리", fontsize=11, labelpad=10)
ax.set_ylabel("감성 라벨", fontsize=11, labelpad=10)

# 격자선 (셀 구분용)
ax.set_xticks(np.arange(-0.5, len(상황), 1), minor=True)
ax.set_yticks(np.arange(-0.5, len(감성), 1), minor=True)
ax.grid(which='minor', color='white', linewidth=2)
ax.tick_params(which='minor', size=0)

plt.tight_layout()
plt.savefig("voc_heatmap.png", dpi=150, bbox_inches='tight')
plt.show()
print("저장 완료: voc_heatmap.png")
```
